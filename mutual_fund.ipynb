```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_funds = pd.read_csv("C:\\Users\\2022\\Desktop\\data science projects\\Dataset\\comprehensive_mutual_funds_data.csv")
df_funds.head()
```

```python
top_performers = df_funds.groupby('scheme_name').agg({'returns_3yr': 'mean'}).sort_values(by='returns_3yr', ascending=False)
print("\nTop Performing Mutual Funds:")
top_performers.head()
```

```python
# Correlation Analysis between numerical columns
correlation_matrix = df_funds[['min_sip', 'min_lumpsum', 'expense_ratio', 'fund_size_cr', 'fund_age_yr', 
                         'risk_level', 'rating', 'returns_1yr', 'returns_3yr', 'returns_5yr']].corr()
print("\nCorrelation Matrix:")
correlation_matrix
```

```python
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df_funds['expense_ratio'], bins=30, kde=True)
plt.title('Distribution of Expense Ratios')
plt.show()
```

```python
plt.figure(figsize=(14, 7))
sns.histplot(df_funds['returns_1yr'], color='blue', label='1 Year Returns', kde=True, alpha=0.6)
sns.histplot(df_funds['returns_3yr'], color='green', label='3 Years Returns', kde=True, alpha=0.6)
sns.histplot(df_funds['returns_5yr'], color='red', label='5 Years Returns', kde=True, alpha=0.6)
plt.legend()
plt.title('Distribution of Returns')
plt.show()
```

```python
top_10_funds = top_performers.head(10).reset_index()
plt.figure(figsize=(12, 8))
sns.barplot(x='returns_3yr', y='scheme_name', data=top_10_funds, palette='viridis')
plt.title('Top 10 Mutual Funds Based on 3-Year Returns')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(x='expense_ratio', y='returns_3yr', data=df_funds)
plt.title('Expense Ratio vs. 3-Year Returns')
plt.show()
```
